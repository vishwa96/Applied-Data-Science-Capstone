```python
def get_pin(page,Area_Name,Pin_Code) :
    wiki_link='http://www.chennaiiq.com/india/pincode/index.asp?id=35&state_name=West%20Bengal&page='+page
    raw_wiki_page_content = requests.get(wiki_link).text
#print(wiki_link)
    raw_wiki_page_content_xml = BeautifulSoup(raw_wiki_page_content,'html.parser')
    table = None
    for tab_cell in raw_wiki_page_content_xml.find_all('table', class_='TBox'):
        table = tab_cell
    line_number = 1
    for tr_cell in table.find_all('tr'):
        if line_number > 1 :
            column_number = 1
            for td_cell in tr_cell.find_all('td'):
                if column_number == 3 :
                #print(td_cell.text)
                    Area_Name.append(td_cell.text)
                if column_number == 5 :
                    Pin_Code.append(td_cell.text)
                column_number = column_number + 1
        line_number = line_number + 1
```

```python
# calculate distance from kolkata for all the areas
import geopy.distance

kol_latitude = 22.5726; kol_longitude = 88.3639 #From Google search
Area_Name = []
Pin_Code = []
Lat = []
Lng = []
Distance = []

for index, row in kolkata_df.iterrows():
    PinCode_temp = row['PinCode']
    AreaName_temp = row['AreaName']
    AreaName_temp = AreaName_temp.strip()
    Lat_temp = row['Latitude']
    Lng_temp = row['Longitude']
    distance = geopy.distance.vincenty((kol_latitude,kol_longitude), (Lat_temp,Lng_temp)).km
    Area_Name.append(AreaName_temp)
    Pin_Code.append(PinCode_temp)
    Lat.append(Lat_temp)
    Lng.append(Lng_temp)
    Distance.append(distance)

kolkata_df = pd.DataFrame({"PinCode":Pin_Code,"AreaName":Area_Name,"Latitude":Lat,"Longitude":Lng,"Distance":Distance})
kolkata_df
```

```python
import json # library to handle JSON files
from pandas.io.json import json_normalize
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```
