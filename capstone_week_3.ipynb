# Creating Dataframe by Scrapping Wikipedia

```python
# iporting librabries
import pandas as pd
import numpy as np
```

```python
# reading data
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', na_values = 'Not assigned', flavor = 'html5lib')
```

```python
print(type(df[0]))
df = df[0]
df.columns
```

```python
df.rename({0: 'Postcode', 1: 'Borough', 2: 'Neighbourhood'}, axis = 'columns', inplace = True)
```

```python
# dealing with missing data
print(df.isnull().sum())
df.drop(df.index[0], inplace = True)
df.dropna(axis = 0, inplace = True)
print(df.isnull().sum())
df
```

```python
# Number of boroughs and neighbourhoods
print('The dataframe has {} boroughs and {} neighbourhoods.'.format(

len(df['Borough'].unique()),

len(df['Neighbourhood'].unique())))
print(df.shape)
```
